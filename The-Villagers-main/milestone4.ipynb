### SQL Query Top 5 bollywoood movies over last 7 years, and their ratings. 

```python
%%bigquery
SELECT b.title, tr.averageRating
FROM datamart.Bollywood b
LEFT JOIN datamart.Title_basics tb on b.titleId = tb.titleId
LEFT JOIN datamart.Title_ratings tr on tb.titleId = tr.titleId
WHERE b.title != "None" and EXISTS(SELECT tb.startYear FROM datamart.Title_basics tb2 WHERE tb2.startYear > 2014 and tb2.startYear < 2021 and tb.startYear = tb2.startYear)
GROUP BY b.title, tr.averageRating
ORDER BY tr.averageRating DESC
Limit 5
```

### Most created Genres - Bollywood vs IMDB in last 7 years, limited to 10

```python
%%bigquery
SELECT bg.genre as BollywoodGenre, b.title, tg.genres as IMDBgenres
FROM datamart.bollywoodGenre_Dataflow bg
LEFT JOIN datamart.Bollywood b on bg.titleId = b.titleId
LEFT JOIN datamart.Title_basics tb on tb.titleId = b.titleId
INNER JOIN datamart.titleGenres_Dataflow tg on tg.titleId = tb.titleId
WHERE b.title != "None" and EXISTS(SELECT tb.startYear FROM datamart.Title_basics tb2 WHERE tb2.startYear > 2014 and tb2.startYear < 2021 and tb.startYear = tb2.startYear)
GROUP BY bg.genre, tg.genres, b.title
Limit 10
```

### What are the top 20 recent (2001-2021) Bollywood movies that have a lead actress and have a high average ratings

```python
%%bigquery
SELECT distinct ba.name as BollywoodActress, b.title, count(b.title) as MovieCount, tb.startYear, tr.averageRating
FROM datamart.Bollywood_actress ba
INNER JOIN datamart.Bollywood b on b.title = ba.debutAsLeadRole
INNER JOIN datamart.Title_basics tb on tb.titleId = b.titleId
INNER JOIN datamart.Title_ratings tr on tr.titleId = tb.titleId
WHERE tb.startYear > 2000 and tb.titleType = "movie"
GROUP BY ba.name, b.title, tr.averageRating, tb.startYear
ORDER By tr.averageRating DESC
Limit 20
```

### Views 

### top 5 Bollywood 

```python
%%bigquery
CREATE VIEW reports.Top_5_Bollywood_Movies AS
    SELECT b.title, tr.averageRating
    FROM datamart.Bollywood b
    LEFT JOIN datamart.Title_basics tb on b.titleId = tb.titleId
    LEFT JOIN datamart.Title_ratings tr on tb.titleId = tr.titleId
    WHERE b.title != "None" and EXISTS(SELECT tb.startYear FROM datamart.Title_basics tb2 WHERE tb2.startYear > 2014 and tb2.startYear < 2021 and tb.startYear = tb2.startYear)
    GROUP BY b.title, tr.averageRating
    ORDER BY tr.averageRating DESC
    Limit 5
```

### Most created genres

```python
%%bigquery
CREATE VIEW reports.Most_Created_Genres AS
    SELECT bg.genre as BollywoodGenre, b.title, tg.genres as IMDBgenres
    FROM datamart.bollywoodGenre_Dataflow bg
    LEFT JOIN datamart.Bollywood b on bg.titleId = b.titleId
    LEFT JOIN datamart.Title_basics tb on tb.titleId = b.titleId
    INNER JOIN datamart.titleGenres_Dataflow tg on tg.titleId = tb.titleId
    WHERE b.title != "None" and EXISTS(SELECT tb.startYear FROM datamart.Title_basics tb2 WHERE tb2.startYear > 2014 and tb2.startYear < 2021 and tb.startYear = tb2.startYear)
    GROUP BY bg.genre, tg.genres, b.title
    Limit 10
```

### Highest rated lead actress Bollywood Movies 

```python
%%bigquery
CREATE VIEW reports.Lead_Actress_Bollywood_Movies AS
    SELECT distinct ba.name as BollywoodActress, b.title, count(b.title) as MovieCount, tb.startYear, tr.averageRating
    FROM datamart.Bollywood_actress ba
    INNER JOIN datamart.Bollywood b on b.title = ba.debutAsLeadRole
    INNER JOIN datamart.Title_basics tb on tb.titleId = b.titleId
    INNER JOIN datamart.Title_ratings tr on tr.titleId = tb.titleId
    WHERE tb.startYear > 2000 and tb.titleType = "movie"
    GROUP BY ba.name, b.title, tr.averageRating, tb.startYear
    ORDER By tr.averageRating DESC
    Limit 20
```
