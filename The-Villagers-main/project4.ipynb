### Create the Firestore collections and subcollections based on your design and populate them with the shopify data.


```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
```

Create Categories collection and Apps subcollection

```python
batch = db.batch()
counter = 0

df = pd.read_csv('/home/jupyter/spanner_shopify/categories.csv', sep=',', header=0, lineterminator='\n')
category_rows = df.values.tolist()

df = pd.read_csv('/home/jupyter/spanner_shopify/apps.csv', sep=',', header=0, lineterminator='\n')
apps_rows = df.values.tolist()

for category_row in category_rows:
    counter += 1

    categories_record = {}
    categories_record['id'] = category_row[0]
    categories_record['title'] = category_row[1]

    category_ref = db.collection('category3').document(category_row[1])

    batch.set(category_ref, categories_record)

for apps_row in apps_rows:
    counter += 1
    apps_record = {}
    apps_record['id'] = apps_row[0]
    apps_record['url'] = apps_row[1]
    apps_record['title'] = apps_row[2]
    apps_record['developer'] = apps_row[3]
    apps_record['developer_link'] = apps_row[4]
    apps_record['icon'] = apps_row[5]
    apps_record['rating'] = apps_row[6]
    apps_record['reviews_count'] = apps_row[7]

    apps_ref = category_ref.collection('apps3').document(apps_row[0])
            
    batch.set(apps_ref, apps_record)

    if(counter > 399):
        counter = 0
        batch.commit()
        batch = db.batch()
    
batch.commit()
```

### Create Apps Collection

```python
batch = db.batch()
counter2 = 0

df = pd.read_csv('/home/jupyter/spanner_shopify/apps.csv', sep=',', header=1, lineterminator='\n')
rows = df.values.tolist()

for row in rows:
    counter2 += 1

    record = {}
    record['id'] = row[0]
    record['url'] = row[1]
    record['title'] = row[2]
    record['developer'] = row[3]
    record['developer_link'] = row[4]
    record['icon'] = row[5]
    record['rating'] = row[6]
    record['reviews_count'] = row[7]

    app_ref = db.collection('apps3').document(row[0])

    batch.set(app_ref, record)

    if(counter2 == 399):
        counter2 = 0
        batch.commit()
        batch = db.batch()
    
batch.commit()
```

### List the 10 apps with the highest number of reviews (based on apps.reviews_count)
Return the id, title, developer, rating and reviews_count of those apps.

```python
query = db.collection('apps3').where('reviews_count','>', 0).order_by('reviews_count','DESCENDING').limit(10)

results = query.stream()
result_dict = {}

for result in results:
    resultRow = result.to_dict()
    result_dict.update(resultRow)
    print("id: "+ str(result_dict['id']) + " title: " + result_dict['title'] + " developer: " + result_dict['developer'] + " rating: " + str(result_dict['rating']) + " reviews_count: " + str(result_dict['reviews_count']))
    
#     print(f'{result.id} => {result.to_dict()} \n')
    
```
