### (IMDB dataset joined to Bollywood dataset) 
### Selected the titles and genres from the IMDB_staging.title_basics and Bollywood_staging.bollywood tables. Then left joined them by movie title name where the genre equals "legend" and not null. Then ordered by Bollywood's movie title. This is to see if IMDB's genres are the same as the Bollywood dataset's 'legend' genre movies.

```python
%%bigquery
select tb.originalTitle, tb.genres, b.Title, b.Genre
from IMDB_staging.title_basics tb left join Bollywood_staging.bollywood b
on tb.originalTitle = b.Title
where b.Genre = "legend" and tb.genres != "\\N"
Order By b.Title
```

## (Bollywood dataset) Movies directors made, group by director, year and title from bollywood having the year group = 1920, order by director, limit 20


```python
%%bigquery
select b.Director, b.Year, b.Title 
from Bollywood_staging.bollywood b 
Group By b.Director, b.Title, b.Year Having b.Year = 1920 
Order By b.Director 
Limit 20 
```

### (IMDB dataset) Query viewing titles that changed from their original title to a new primary title, removing null runtime and start year entries. Grouped by runtime, original title, primary title, start year, limited to 20, ordered by start year. 

```python
%%bigquery
select tb.originalTitle, tb.primaryTitle, tb.startYear, tb.runtimeMinutes
from IMDB_staging.title_basics tb 
where tb.originalTitle != tb.primaryTitle AND tb.startYear != '\\N' AND tb.runtimeMinutes != '\\N'  and tb.titleType = "movie"
Group By tb.originalTitle, tb.primaryTitle, tb.startYear, tb.runtimeMinutes 
Order By tb.startYear DESC 
Limit 20
```

### (Bollywood dataset joined to IMDB dataset by two tables) 
### Selected the title, genre, start year, and average rating from Bollywood_staging.bollywood left joined to IMDB_staging.title_basics on the movie title and right joined to IMDB_staging.title_ratings on the title id, 'tconst'. Where the tb.titleType = movie and tb.genres not equal to null and tr.averageRating > 9. Ordered by the start year and average rating (in descending order) limited by 10 records.
### What genre of Bollywood movies had the highest ratings for each recent year?

```python
%%bigquery
select b.Title, b.Genre, tb.startYear, tr.averageRating
from Bollywood_staging.bollywood b 
left join IMDB_staging.title_basics tb
on b.Title = tb.originalTitle
right join IMDB_staging.title_ratings tr
on tb.tconst = tr.tconst
where tb.titleType = "movie" and tb.genres != "\\N" and tr.averageRating > 9
Order By tb.startYear DESC, tr.averageRating DESC
Limit 10;
```

### (Bollywood dataset joined to IMDB dataset by three tables) 
### Selected the title, genre, averageRating, and language from Bollywood_staging.bollywood on a left join on the movie title. Then a right join on IMDB_staging.title_ratings on tconst and then left joined on IMDB_staging.title_akas on the titleId where the titleType = 'movie' and the genre = 'historical' and language = english and not null. Then grouped by the Title, Genre, averageRating, and language as well as ordered by averageRating in descending order.
### Are there alot of historical bollywood movies in english and are they highly rated?

```python
%%bigquery
select b.Title, b.Genre, tr.averageRating, ta.language
from Bollywood_staging.bollywood b 
left join IMDB_staging.title_basics tb
on b.Title = tb.originalTitle
right join IMDB_staging.title_ratings tr
on tb.tconst = tr.tconst
left join IMDB_staging.title_akas ta
on tr.tconst = ta.titleId
where tb.titleType = "movie" and b.Genre = "historical" and ta.language != "\\N" and ta.language = "en"
GROUP BY b.Title, b.Genre, tr.averageRating, ta.language
Order By tr.averageRating DESC
```

### Select primary title, genre, year from title basics table and rating, number of votes from title rating table. Joined tables at title ID, tconst. Grouped by number of votes, title, genre average rating ans start year. Ordered by number of votes in descending order, limited to 20

### From 1995-2010, which movies/genres had the highest amount of votes?   

```python
%%bigquery
select tb.primaryTitle, tb.genres, tr.averageRating, tb.startYear, tr.numVotes from IMDB_staging.title_basics tb join IMDB_staging.title_ratings tr on tb.tconst = tr.tconst 
where tb.startYear > '1995' AND tb.startYear < '2010' AND tb.genres != '\\N'
GROUP BY tr.numVotes, tb.primaryTitle, tb.genres, tr.averageRating, tb.startYear
Order By tr.numVotes DESC
Limit 20
```

### IMBD: joining title principle and name basics for actress name and joining title basics to title principal for movie title. Grouped by primary title, category and name. Ordered by primary name in descending order, limited to 20.  

### Which titles have an actress in the principal cast?

```python
%%bigquery 
select tp.category, nb.primaryName, tb.primaryTitle
from IMDB_staging.title_principals tp join IMDB_staging.name_basics nb on tp.nconst = nb.nconst join IMDB_staging.title_basics tb on tb.tconst = tp.tconst  
where tp.category = 'actress'
Group By tb.primaryTitle, tp.category, nb.primaryName
Order By nb.primaryName DESC
Limit 20 
```

### Bollywood: selected name, debut as lead role, title, year and highest gross earned from actress table and bollywood table. Right joined actress table to bollywood table on title. In range between 1950 and 2020. Ordered by highest gross earned in descending order, limited to 10. 

### Bollywood: Highest grossing movie with a female lead from 1950 to 2020?

```python
%%bigquery 
select ba.Name, ba.Debut_aslead_role, b.Title, b.Year, b.Highest_Grosser_By_Year_in_crores_
from Bollywood_staging.bollywood b 
right join Bollywood_staging.bollywood_actress ba on b.Title = ba.Debut_aslead_role 
where b.Year >1950 and b.Year < 2020 
Order By b.Highest_Grosser_By_Year_in_crores_ DESC
LIMIT 10
```

### Bollywood: select bollywood actress name, height, title of movie, year, gross earnings of movie and join to bollywood table at title. Where gross earned is above 0, ordered by movie gross earnings in descending order, limited to 20.

### Do higher grossing movies prefer taller women or shorter women?

```python
%%bigquery 
select ba.Name, ba.Debut_aslead_role, ba.Height_in_cm_, b.Year, b.Highest_Grosser_By_Year_in_crores_
from Bollywood_staging.bollywood b 
right join Bollywood_staging.bollywood_actress ba on b.Title = ba.Debut_aslead_role 
where b.Highest_Grosser_By_Year_in_crores_  > 0.0
Order By b.Highest_Grosser_By_Year_in_crores_  DESC
LIMIT 20
```

### IMDB: Do people like short or long movies, tell by average rating

```python
%%bigquery 
select tb.titleType, tr.averageRating, tb.primaryTitle
from IMDB_staging.title_ratings tr right join IMDB_staging.title_basics tb on tr.tconst = tb.tconst  
where tb.titleType = 'short' OR tb.titleType = 'movie' AND tr.averageRating > 0
Group By tr.averageRating, tb.primaryTitle, tb.titleType
Order By tr.averageRating DESC
Limit 20
```
